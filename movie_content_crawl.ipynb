```python
import requests
import bs4
import requests, bs4
import csv

total_img_link = []
with open('yahoo_movie.csv', 'r', encoding='utf-8-sig') as f:
    result = f.readlines()

for i in range(1, len(result)):
    try:
        movie_url = result[i].split(',')[2]
        # print(movie_url)
        moviehtml = requests.get(movie_url)
        objSoup = bs4.BeautifulSoup(moviehtml.text, 'html')
        movie_img_link = objSoup.select("div.movie_intro_foto")[0].find("img").get("src")
        print(movie_img_link)
        total_img_link.append(movie_img_link)
    except:
        movie_img_link = 'https://www.portent.com/images/2020/07/404-Error-Image.jpg'
        total_img_link.append(movie_img_link)
        print('無圖片')


with open('img_link.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['link'])
    for link in total_img_link:
        writer.writerow([link])
f.close()
```

```python
url='https://www.truemovie.com/tairelease.htm'
```

```python
moviehtml = requests.get(url)
objSoup = bs4.BeautifulSoup(moviehtml.text, 'html')
```

```python
movie_link=objSoup.find_all("a")
```

```python
objSoup.find_all("a")[34]["href"]
```

```python
urls=[]
for i in range(35,len(movie_link)):
    url="https://www.truemovie.com/{}".format(objSoup.find_all("a")[i]["href"])
    print(url)
    urls.append(url)
    
len(urls)
```

```python
with open('url_link.csv','w',newline='') as f:
        writer=csv.writer(f)
        writer.writerow(['url'])
        for url in urls:
            writer.writerow([url])
f.close()
```

```python
import time
with open('movie2021.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['movie_name','movie_type','movie_content'])
for url in urls[:]:
    try:
        res = requests.get(url)
        res.text
        res.encoding='Big5'
        objSoup = bs4.BeautifulSoup(res.text, 'lxml')
        i=0
        for i in range(len(objSoup.find_all("font"))):
            if objSoup.find_all("font")[i].text[:4]=='中文片名':
                movie_name=objSoup.find_all("font")[i+1].text
                print(movie_name)
            elif objSoup.find_all("font")[i].text[:8]=='\n\n\n\n中文片名':
                movie_name=objSoup.find_all("font")[i].text.split('：')[1]
                print(movie_name)
            if objSoup.find_all("font")[i].text[:2]=='類型':
                movie_type=objSoup.find_all("font")[i].text[3:]
                print(movie_type)
            if objSoup.find_all("font")[i].text[:4]=='劇情簡介':
                movie_content=objSoup.find_all("font")[i].text[5:]
                movie_content=movie_content.replace('\n','')
                movie_content=movie_content.replace('\t','')
                print(movie_content)
        print('\n')
        with open('movie2021.csv','a+',newline='',encoding='utf-8') as f:
            writer=csv.writer(f)
            writer.writerow([movie_name,movie_type,movie_content])
        time.sleep(1)
    except:
        pass
f.close()
```
